```python
from PIL import Image
from pathlib import Path
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import csv
```

```python
df = pd.read_csv('dataframe.csv', index_col=0)
df
```

```python
folder_path = '10x'

for file_idx, file in enumerate(Path(folder_path).glob('*.png')):
    pic = Image.open(file)
    f_name = os.path.splitext(file.name)[0]
    as_array = np.array(pic)[...,2].T
    tissue = np.argwhere(as_array<240)
    vessels = np.argwhere(as_array>240)
    tissue_pixels = tissue.shape[0]
    vessel_pixels = vessels.shape[0]
#     print("Number of pixels:", count)
    df.loc[df['name'] == f_name, 'tissue_pixels'] = tissue_pixels
    df.loc[df['name'] == f_name, 'vessel_pixels'] = vessel_pixels

df.to_csv('updated_dataframe.csv', index=False)
```
